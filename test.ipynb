```python
csv_file_path = os.getcwd() + '/bees.csv'
bee_folder = os.getcwd() + "/images/bees_final/" 
labels_dict = {'honeybee': 0,
'bumblebee': 1,'carpenterbee': 2}
#Bees DataFrame
df = pd.read_csv(csv_file_path)
#Bees Label json
label = df['Label']

#display(df)
```

```python

#for i in range(len(df)):
for i in range(0,10):
    '''Skip images that have no bee labels'''
    if bool(df['Skipped'][i]) is True:
        print('Skipped bee_image' + str(i))
        continue

    '''Create file name'''
    prefix = 'bees_image'
    file_number = str(i)
    image_file_name = prefix + file_number + ".jpg"
    bee_image = bee_folder + image_file_name

    '''Download image from url and put in bees folder'''
    f = open(bee_image,'wb')
    f.write(requests.get(df['Labeled Data'][i]).content)
    f.close()
    print('Downloaded ' + image_file_name)

    '''Get Image size'''
    image = Image.open(bee_image)
    width, height = image.size #get total image size

    '''Get bee label and store coords in txt file'''
    bee_label(i)
    print("Created " + prefix + file_number + '.txt')

```

```python
path_list = []
for images in os.listdir(bee_folder):
    if (images.endswith(".jpg")):
        path_list.append(images)

from sklearn.model_selection import train_test_split
train_data, test_data = train_test_split(path_list, test_size=0.1)
```

```python
def create_file(name, data):
    f = open(os.getcwd() + "/images/"+ name + ".txt", "w")
    for path in data:
        f.write('/content/gdrive/My Drive/darknet/img/' + path + '\n')
    print("Created:", name + '.txt')
    f.close()

create_file('test', test_data)
create_file('train', train_data)
```
